```python
import datetime
from datetime import date
from pandas.tseries.offsets import *
import pandas as pd
```

```python
# lagging with one month gap
offset= 12
counter = 0
for year in range(2012,2019):
    for month in range(1,13):
        d = date(year,month,1)
        d3 = d+DateOffset(months=-1)
        d2 = d+DateOffset(months=-(offset+1))
        if counter == 0:
            print("drop table if exists lagging_feature_gap_%d;\n" % offset)
            print('create table if not exists lagging_feature_gap_%d as' % offset)
        else:
            print('insert into lagging_feature_gap_%d' % offset)
        print("select province_id, city_id, class_id, to_date('%s','yyyymmdd') as sale_date, avg(sale_quantity) as avg_%d,median(sale_quantity) as med_%d, max(sale_quantity) as max_%d, min(sale_quantity) as min_%d,stddev(sale_quantity) as std_%d, count(sale_quantity) as count_%d" \
             %(d.strftime(format='%Y%m%d'), offset,offset,offset,offset,offset,offset))
        print('from yc_car_sales_grouped a')
        print("where a.sale_date<to_date('%s','yyyymmdd') and a.sale_date>=to_date('%s','yyyymmdd')"%(d3.strftime(format='%Y%m%d'),d2.strftime(format='%Y%m%d')))
        print("group by province_id, city_id, class_id;\n")
        counter+=1
        
        if year==2018 and month==2:
            break
```
